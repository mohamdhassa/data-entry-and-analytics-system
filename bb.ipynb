```python
import sys
print(sys.executable)

import pandas as pd
import matplotlib.pyplot as plt
import os
import IPython

# Check if the notebook is running in a specific environment before using environment-specific variables
if hasattr(IPython, 'extract_module_locals'):
    try:
        notebook_name = "/".join(
            IPython.extract_module_locals()[1]["__vsc_ipynb_file__"].split("/")[-5:]
        )
        print(notebook_name)
    except KeyError:
        print("Variable '__vsc_ipynb_file__' not found.")
else:
    print("Not running in a VS Code environment.")

# Load the dataset
try:
    df = pd.read_csv("output.csv")
except FileNotFoundError:
    print("Error: 'output.csv' file not found. Please check the file path.")
    exit()
except pd.errors.ParserError:
    print("Error: Unable to read the 'output.csv' file. Please check the file format.")
    exit()

# Check if the DataFrame is empty
if df.empty:
    print("Error: The DataFrame is empty. Please check the contents of the 'output.csv' file.")
    exit()

# Calculate churn rate by gender
gender_churn = df.groupby('gender')['churn'].mean()

# Plot pie chart for churn rate by gender
fig, ax = plt.subplots()
ax.pie(gender_churn, labels=gender_churn.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Churn Rate by Gender')

# Save the plot
file_path = 'C:\\Users\\USER\\Desktop\\nn\\nn\\public\\output2.png'
plt.savefig(file_path)

# Show Plot
plt.show()
```
